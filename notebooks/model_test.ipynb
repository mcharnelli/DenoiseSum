```python
import os

import numpy as np
import torch
import torch.nn as nn
#from torch.utils.data.dataset import BufferedShuffleDataset
from tqdm.auto import tqdm
from DenoiseSum.LanguageModel.data import LanguageModelData

from DenoiseSum.LanguageModel.lm import LM
from DenoiseSum.utils import get_lm_dict, pad, Batcher

import pickle
import os.path
from DenoiseSum import DATA_PATH
```

```python
dataset = 'chisito'
chisito_path = DATA_PATH / 'chisito' 
model_file = 'lm.model'
model_file = DATA_PATH / 'model' / dataset / model_file 
with open(chisito_path / 'reviews_small_word_dict.pickle', "rb") as handle:
    word_dict = pickle.load(handle)

word_size = len(word_dict)
word_dim = 300
hidden_dim = 512
model = LM(word_size, word_dim, hidden_dim)
optimizer = torch.optim.Adagrad(
    model.parameters(), lr=0.1, initial_accumulator_value=0.1
)
best_point = torch.load(model_file)
model.load_state_dict(best_point["state_dict"])
optimizer.load_state_dict(best_point["optimizer"])
```

```python
word_dict_inv = {i:word for word, i in word_dict.items()}
```
